```python
import airportsdata

# Load the airports database (IATA codes)
airports = airportsdata.load('IATA')

def get_lat_lon(iata):
    info = airports.get(iata)
    if info:
        return pd.Series({'lat': info['lat'], 'lon': info['lon']})
    else:
        return pd.Series({'lat': None, 'lon': None})

# Create df_coordinates from df_airports_df
df_coordinates = df_airports_df.copy()

# Get departure coordinates
dep_coords = df_coordinates['departure_point'].apply(get_lat_lon)
df_coordinates['dep_lat'] = dep_coords['lat']
df_coordinates['dep_long'] = dep_coords['lon']

# Get arrival coordinates
arr_coords = df_coordinates['arrival_point'].apply(get_lat_lon)
df_coordinates['arr_lat'] = arr_coords['lat']
df_coordinates['arr_long'] = arr_coords['lon']

# Fill SXF manually if still missing
sxf_lat, sxf_long = 52.380001, 13.522500
df_coordinates.loc[df_coordinates['departure_point'] == 'SXF', ['dep_lat', 'dep_long']] = [sxf_lat, sxf_long]
df_coordinates.loc[df_coordinates['arrival_point'] == 'SXF', ['arr_lat', 'arr_long']] = [sxf_lat, sxf_long]

df_coordinates
```

```python
df_coordinates = pd.DataFrame(df_coordinates)
```

```python
df_coordinates.isnull().sum()
```

```python
import numpy as np

# Haversine formula to calculate distance in km
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in kilometers
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

# Calculate the distance for each row
df_coordinates['distance_km'] = haversine(
    df_coordinates['dep_lat'],
    df_coordinates['dep_long'],
    df_coordinates['arr_lat'],
    df_coordinates['arr_long']
)
```

```python
# Drop duplicates for merge keys
df_coordinates_unique = df_coordinates.drop_duplicates(subset=['departure_point', 'arrival_point'])

# Merge only the unique pairs
df_maindf = df_maindf.merge(
    df_coordinates_unique[['departure_point', 'arrival_point', 'distance_km']],
    on=['departure_point', 'arrival_point'],
    how='left'
)
```

```python
df_maindf['distance_km'] = (df_maindf['distance_km'].round(2) * 100).fillna(0).astype(int)
```
